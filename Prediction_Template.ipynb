```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
import warnings

warnings.filterwarnings('ignore')
```

# Data Collection

```python
# input year and playertype
# get list of dataframes
def get_season_data_pfr(year, playertype):
    # get page
    page = requests.get(f'https://www.pro-football-reference.com/years/{year}/{playertype}.htm')
    # read content for tables with pandas
    return pd.read_html(page.content)
```

# Data Processing

## Data Cleaning

```python
def remove_toplevel(df):
    df.columns = df.columns.droplevel(0)
    return df
# gets rid of the string rows
def remove_stringrows(df):
    df = df[df['Player']!='Player']
    return df

# takes the * and + out of names
def remove_asterixplus(df):
    df = df[df['Player']!='Player']
    return df
```

```python
#pfr_r2020 = remove_toplevel(pfr_r2020)
#pfr_r2021 = remove_toplevel(pfr_r2021)
#pfr_r2022 = remove_toplevel(pfr_r2022)

pfr_r2020 = remove_stringrows(pfr_r2020)
pfr_r2021 = remove_stringrows(pfr_r2021)
pfr_r2022 = remove_stringrows(pfr_r2022)

pfr_r2020 = remove_asterixplus(pfr_r2020)
pfr_r2021 = remove_asterixplus(pfr_r2021)
pfr_r2022 = remove_asterixplus(pfr_r2022)
```

```python
# joins the data
pfr_merged = pfr_r2020.join(pfr_r2021, lsuffix='', rsuffix='_2021')
pfr_merged = pfr_merged.join(pfr_r2022, rsuffix='_2022')

# remove no data found 
pfr_merged = pfr_merged.dropna()
```

```python
pfr_merged[['Yds', 'Yds_2021', 'Yds_2022']]
```

```python
pfr_merged['Yds'] = pfr_merged['Yds'].astype(float)
pfr_merged['Yds_2021'] = pfr_merged['Yds_2021'].astype(float)
pfr_merged['Yds_2022'] = pfr_merged['Yds_2022'].astype(float)
```

```python
pfr_merged['Att'] = pfr_merged['Att'].astype(float)
pfr_merged['Att_2021'] = pfr_merged['Att_2021'].astype(float)
pfr_merged['Att_2022'] = pfr_merged['Att_2022'].astype(float)
```

```python
# plot
plt.scatter(pfr_merged['Yds'], pfr_merged['Yds_2021'])
```

```python
# split into features and response
# X is feature
X = pfr_merged['Yds'].values.reshape(-1,1) # reshape for single feature
# y is response
y = pfr_merged['Yds_2021']
```

```python
# get predictions
# now we use 2021 data to predict 2022, using the same model 
X = pfr_merged['Yds_2021'].values.reshape(-1,1) # reshape for single feature
y_pred = model.predict(X)
```

```python
y_true = pfr_merged['Yds_2022']
```

```python
# look at how well the model does
# r^2 tells you what % of the variance in your target variable is explained by your features
r2 = r2_score(y, y_pred)
r2
```
